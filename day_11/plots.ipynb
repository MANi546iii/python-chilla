```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
kashti = sns.load_dataset('titanic')
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived',hue = 'sex',data = kashti)
```

```python
# scatter plot 
plt.scatter( data = kashti, x='age',y='fare')
```

```python
sns.scatterplot(data = kashti , x ='fare',y ='age',hue = 'pclass')
```

```python
sns.scatterplot(data = kashti , x ='sex',y ='fare',hue = 'pclass')
```

```python
sns.barplot(data = kashti , x ='sex',y ='fare',hue = 'pclass')
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins = 10,figsize = (9,7), grid = False)
```

```python
kashti.columns
```

```python
sns.boxplot(data = kashti , x ='sex',y ='fare',hue = 'who')
```

```python
sns.boxenplot(data = kashti , x ='sex',y ='fare',hue = 'who')
```

```python
sns.stripplot(data = kashti , x ='sex',y ='fare',hue = 'who')
```

```python
#heat plot
corr = kashti.corr()
plt.figure(figsize=(10.10))
sns.heatmap(corr)
```

```python
sns.violinplot(data = kashti , x ='sex',y ='fare',hue = 'who')
```
