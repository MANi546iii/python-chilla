```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

# 1- Import Data

```python
# import dataset
df=sns.load_dataset('titanic')
df
```

# 2- Bigger picture of Data

```python
df.dtypes
```

```python
df.info()  # found missing values
```

```python
# count null value
df.isnull().sum() #find missing values
```

```python
mean_age = df['age'].mean()
df['age']=df['age'].fillna(mean_age)
df
```

```python
df1= df.drop(columns=['deck'])
```

```python
df1
```

```python
df1['embark_town'].fillna(df1['embark_town'].mode()[0] , inplace= True)
df1['embarked'].fillna(df1['embarked'].mode()[0]  , inplace = True)
df1
```

```python
df1.isnull().sum()
```

```python
print('mean =',df1['age'].mean())
print('median = ',df1['age'].median())
print('mode = ', df1['embark_town'].mode())
```
